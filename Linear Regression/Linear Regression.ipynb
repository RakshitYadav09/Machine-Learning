```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.area,df.price, color='red', marker='+')
plt.xlabel('area')
plt.ylabel('price')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
sample = 5000
sample_reshaped = np.array(sample).reshape(1, -1)
reg.predict(sample_reshaped)
```

```python
reg.intercept_
```

```python
obs = 5000
obs_as_matrix = np.reshape(obs, (1,1))
reg.predict(obs_as_matrix)
```

```python
d=pd.read_csv("areas.csv")
```

```python
p=reg.predict(d)
```

```python
d['prices']=p
```

```python
d.head
```

```python
d.to_csv("prediction.csv")
```
